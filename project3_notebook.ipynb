### Introduction

The RUNX1 transcription factor is a part of a family of proteins that are known for being master regulators. Their well defined behavior has inherently tied them into playing important roles in cancer development. RUNX1 specifically has been strongly associated with breast cancer, where it can act as both a tumor suppressor and an oncogene. RUNX1 acts through the alteration of chromatin structure alongside other chromatin modifiers and remodeling enzymes. The study in question specifically aimed to investigate the relationship between RUNX1 mediated transcription and the resulting genome organization. As changes in nuclear architecture are often observed in cases of breast cancer alongside the overexpression of RUNX1, investigating this relationship will help make a more complete picture into breast cancer development. To approach this problem, the authors used a variety of different methods with two conditions, cells with a suppression RUNX1 and cells with no suppression of RUNX1. First, the authors employed a Hi-C approach in order to examine both inter- and intra-chromosomal interactions between both conditions. Next, they utilized RNA-seq analysis in order to compare the gene expression profile of these two conditions. In addition to this, the authors performed a ChIP-seq analysis in order to investigate RUNX1-mediated regulation of chromatin organization through its binding patterns on the chromatin.

### Methods 

To begin this analysis, single end reads were obtained from a ChIPseq experiment that was done in a well established breast cancer line, MCF-7. The quality of these reads were first analyzed using FastQC v0.12.1 with default parameters [1]. At the same time, the GRCh38 human reference genome was indexed using default Bowtie2 v2.5.4 parameters [2]. The adapters and low quality bases were trimmed from the reads using the single end command with default parameters from Trimmomatic v0.40 [3]. Upon completion, the reads were aligned to the previously indexed human reference genome using Bowtie2 v2.5.4 with default parameters [4]. Further quality control was then completed using default MultiQC v1.25 [5] and the default flagstat utility from Samtools v1.22.1 [6]. These alignment files were then sorted in genomic order based on their alignment to the reference genome and indexed using Samtools v1.22.1 with default parameters for further analysis [7].
<br>

Bigwig files were generated from the previously sorted BAM files using default deeptools v3.5.5 and signal coverage plots were calculated using deeptools’s computeMatrix utility [8].  These plots included 2000 bp cushions on either side of the transcription start site and end site.A spearman correlation analysis was also performed with deeptools v3.5.5 utilities plotcorrelation and multibwsummary with default parameters to determine the similarity between samples [9]. 
<br>

Peak calling was executed using the HOMER suite v4.11 using the “-style factor” option [10]. To create one set of reproducible peaks, the peaks were intersected using a simple intersection method using the intersect utility of bedtools v2.31.1 [11]. Any peaks that overlapped in undesirable regions as denoted by the hg38-blacklist v2 as poised by the Boyle lab (https://github.com/Boyle-Lab/Blacklist/tree/master/lists) were removed through the default use of bedtools v2.31.1 [12]. Peak annotation and motif finding was done on the set of filtered peaks using the HOMER suite v4.11 with default options [13]. These peaks were lastly analyzed using the online web tool ENRICR in order to identify any associated biological processes [14]. 



### Justification for Methods Choices

1. When plotting the distances of the correlation coefficients of each sample to determine the similarity of our samples, I chose to utilize a spearman correlation. This is because I thought that the data might have been monotonic and not linear due to the technical variability that is often seen in ChIP-seq experiments through things like sequencing depth and immunoprecipitation  efficiency between samples.

2. When producing a set of reproducible peaks with bedtools intersect, I employed the simple intersection method. I chose this method as it only retains peaks that overlap in both replicates and boasts a higher confidence despite being the most stringent of the strategies. 

### Quality Control Evaluation 

Looking through the MultiQC report shows that the data gathered from the experiment is usable for downstream analysis. First, the general statistics report shows that 3 of the 4 samples have a mapped read depth of around 30 million and 1 sample has a mapped read depth of around 10 million. Consulting the original Fastq files reveals that this dramatic decrease in read depth is present before anything was done with the samples. Further, the Samtools flagstat metrics report that each sample had an alignment rate of the reads to the reference genome at higher than 90%, which is a good sign when compared to the standard alignment of 70%. Looking at the results of Trimmomatic shows that the vast majority of reads in each sample survived, as less than 400,000 reads were dropped for each sample, which is also a good sign. 

Examining the FastQC reports shows some potential issues. First, the report has flagged that many reads from the IP samples contain a very high percentage of duplicate reads. The sequence duplication levels shows that 80% of both INPUT samples have a duplication level of about 1, while 80% of IP replicate 1 has a duplication level of over 10. There is also a noticeable change in the GC content where both of the INPUT samples have a GC content of about 40% and both of the IP samples have a GC content of about 50%. These observations could be potentially expected in context of the experiment, which is a ChIP seq experiment. As this was a binding experiment, the IP samples represent the sequences that the RUNX1 protein has bound to while the INPUT samples represent background noise. These identities would imply that we should see high duplication levels in the IP samples and distinct profiles between the two conditions. Other metrics of the data support this. First two overrepresented sequences were identified, each amounting to be 0.9312% and 0.6192% of all reads. As they are not a significant amount of all reads, this should not be an issue and can be ignored. Additionally, the adapter content graph indicates that some of the raw reads have adapter contamination, which should be removed through the use of trimmomatic. Lastly, the mean quality scores look suitable across all samples, as the phred scores are all above a score of 30 and remain relatively stable. The data may not be the highest quality and could be improved by re-completing the experiment in order for IP replicate 2 to have a comparable number of reads when compared to the other samples. Taking the other factors into account, however, indicates that the data is of acceptable quality for downstream analysis.

A signal coverage plot conveys information regarding the number of sequencing reads that align to specific positions along a reference genome or targeted region to visualize depth and distribution of sequencing data. The plots associated with the IP samples in this ChIP experiment show a sharp, tall peak near the transcription start site (TSS) mark and no peak near the transcription end site (TES). This suggests that the RUNX1 protein binds promoter regions and is more associated with transcription initiation as opposed to transcription elongation or termination. The plots associated with the INPUT samples both show little to no consistent TSS enrichment as in the IP samples. Due to this, the INPUT samples help to validate the results obtained from the IP samples.

The table displaying read information that was reported in the paper shows a large difference in mapped reads when compared to the table that was synthesized with data from the analysis in this report. This report boasts a much higher mapped read rate overall than the paper does. In sample INPUT Replicate 1, the paper reports that they were able to get 19,047,821 reads mapped while the same sample in this report was able to get 28,253,592 mapped reads. INPUT replicate 2 in the paper achieved only 7,699,867 mapped reads while this analysis achieved 10,069,440 mapped reads. IP replicate 1 from the paper was able to get 25,949,243 reads mapped and this analysis was able to get 27,969,494 reads mapped. Finally, IP replicate 2 in the paper only mapped 22,957,545 reads while this analysis was able to map 28,351,783 reads. Despite these differences in mapped reads, the raw reads for each sample was the exact same in the paper and this analysis. There are multiple reasons why these differences can exist. First, the authors could be using different versions of the tools used during the analysis. Between updates, specific tools could have undergone updates that changed the algorithm in a way that may produce different results when compared to older versions. Second, the authors didn’t report which parameters they used for their alignment. It could be likely that they used parameters that they didn’t disclose that resulted in a difference in mapped reads.

The correlation plot done in the paper was a Pearson correlation plot that shows that the IP and INPUT samples are more closely related to themselves and show distinction between the two sample types. These samples are grouped together by the side clustering and the heatmap shows two distinct clusters marked by deep red blocks in the areas on the plot where the samples are plotted against themselves. The correlation plot done in this analysis was a Spearman correlation plot and shows different metrics than the one in the paper. This plot shows that the INPUT samples were more closely related to themselves and distinct from the IP samples. This is shown by the clustering pattern on the side clustering where the INPUT samples were clustered together and the IP samples were not. Additionally, the heatmap shows that there is only one cluster marked with red colors, and therefore higher correlation numbers, in the area correlating with the INPUT. The region with the IP samples shows only moderate correlation, showing a more yellow color and therefore lower correlation values, despite the dark red diagonal. From their Pearson correlation plot, the authors gathered that their experiment was a success. Even though it’s tough to compare Pearson correlation and Spearman correlation plots, the Spearman plot gives more insight into replicate variability. For example, if the experiment truly captured the binding behavior of RUNX1, the IP regions would be all of the sequences where it bound. These samples would be both similar to each other as well as being distinct from the INPUT values. Even though the IP samples were not red colored or given a higher number like the INPUT samples were, they were still moderately similar to each other and distinct from the other. This would suggest that the experiment would have been successful but had some technical variability between replicates.

The venn diagrams synthesized from the analysis done in the paper and the analysis done throughout this report shows that the number of peaks found in both replicate 1 and replicate 2 are reasonably similar and that the number of peaks from each replicate noticeably differ. First, the paper reports that there were 3,466 peaks that were found in both replicates while my analysis reported that there were 3,735 peaks. However, the authors report that they found 3,983 peaks in replicate 1 and 10,465 peaks in replicate 2 while the analysis done in this report shows  13,258 peaks found in replicate 1, 12,098 peaks in replicate 2. There could be multiple reasons for these discrepancies in the number of called peaks. First, the authors used the hg19 reference genome as opposed to the hg38, which is what the analysis in this report uses. These reference genomes differ by thousands of basepairs, which can easily give different results. Further, this analysis was able to map many more reads than the original paper. These additional reads could have had many of the peaks that were identified in this report that were not identified in the paper.

### Analyze the annotated peaks


<img src="pics/ENRICHR.png" alt="Figure S2C" width="600">

To perform this enrichment, gene names were collected from the file that contains the reproduceable, HOMER-annotated peaks that were then ran through the online gene enrichment tool, ENRICHR.  The results that were obtained from this search made logical sense when compared to the subject of the paper as many of the hits were consistent with RUNX1’s known biology in breast cancer. The most common hit was related to the estrogen signaling pathway, which is known to stimulate growth of ER-positive breast cancer tumors through both genomic and non-genomic pathways, aligning with RUNX1’s role as an estrogen receptor co-regulator in MCF-7 cells. This hit was found in the KEGG 2021 Human Pathway, MSigDB Hallmark 2020 Pathway, and the Elsevier Pathway Collection. Other common hits included items related to cell adhesion and migration as well as PI3K/AKT relevant signaling pathways, such as PTEN and PKB, which play a role in tumor suppressor functions. These results are interesting because they help to form a preliminary hypothesis that RUNX1 may coordinate both proliferative and growth-suppressive processes in MCF-7 cells, aligning to its context-dependent role in breast cancer.
